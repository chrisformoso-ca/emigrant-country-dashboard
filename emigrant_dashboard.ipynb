```python
import streamlit as st
import pandas as pd
import plotly.express as px

# Load the cleaned data
file_path = '/content/drive/MyDrive/Colab Notebooks/Emigrants by Country 1981-2022/Cleaned_Emigrant_Data.xlsx'
df = pd.read_excel(file_path)

# Streamlit app implementation (same as discussed)
st.title('Emigrant Data Analysis (1981-2022)')
st.subheader('Raw Data')
st.write(df.head())

country = st.selectbox('Select a Country', df['Country'].unique())
filtered_data = df[df['Country'] == country]
st.subheader(f'Emigrant Count Over Time - {country}')
fig = px.line(filtered_data, x='Year', y='Emigrant_Count', title=f'Emigrant Count from {country} Over Time')
st.plotly_chart(fig)

year = st.slider('Select Year', int(df['Year'].min()), int(df['Year'].max()), step=1)
yearly_data = df[df['Year'] == year]
st.subheader(f'Emigrant Count Across Countries in {year}')
fig2 = px.bar(yearly_data, x='Country', y='Emigrant_Count', title=f'Emigrant Count Across Countries in {year}')
st.plotly_chart(fig2)

st.subheader('Descriptive Statistics')
st.write(df.describe())

st.subheader('Download Filtered Data')
st.download_button(
    label="Download data as CSV",
    data=filtered_data.to_csv(index=False),
    file_name=f'{country}_emigrant_data.csv',
    mime='text/csv',
)
```
