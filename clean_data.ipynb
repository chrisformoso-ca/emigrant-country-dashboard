```python
import pandas as pd

# Load the dataset
file_path = '/content/drive/MyDrive/Colab Notebooks/Emigrants by Country 1981-2022/Emigrant-1981-2022-MajorCountry.xlsx'
df = pd.read_excel(file_path, header=None)

# Select specific rows and columns, then combine and clean headers
df = df.iloc[2:46, :12]
headers = df.iloc[0].fillna('') + ' ' + df.iloc[1].fillna('')
headers = headers.str.strip().str.replace(' ', '_')
df.columns = headers
df = df.iloc[2:].reset_index(drop=True)
df.rename(columns={df.columns[0]: 'Year'}, inplace=True)

# Reshape the data to long format
df_long = df.melt(id_vars=['Year'], var_name='Country', value_name='Emigrant_Count')

# Convert 'Year' column to numeric, ensure 'Emigrant_Count' is integer, and fill in nulls/blanks with 0
df_long['Year'] = pd.to_numeric(df_long['Year'], errors='coerce')
df_long['Emigrant_Count'] = df_long['Emigrant_Count'].fillna(0).astype(int)
df_long = df_long.reset_index(drop=True)

# Display the resulting DataFrame and save it to a new Excel file
df_long.info()
print(df_long.head())
df_long.to_excel('/content/drive/MyDrive/Colab Notebooks/Emigrants by Country 1981-2022/Cleaned_Emigrant_Data.xlsx', index=False)
```
